```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2 = sns.load_dataset("taxis")
df2
```

```python
df_total=df2[["passengers", "distance", "fare", "tip", "tolls", "total"]]
df_total
df_total.info()
```

```python
df_total.corr()
```

```python
ax = sns.heatmap(df_total.corr(), vmin=-1, vmax=1, cmap="mako");
```

```python
ax = sns.scatterplot(data=df_total, x='fare', y='distance')
coef = df_total['fare'].corr(df_total['distance'])
plt.text(0.07,0.92, f'Coeficiente de correlacion de pearson = {coef:.2f}')
```

```python
ax = sns.scatterplot(data=df_total, x='distance', y='fare')
coef = df_total['distance'].corr(df_total['fare'])
plt.text(0.90,0.97, f'Coeficiente de correlacion de pearson = {coef:.2f}')
```
