```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df
```

```python
ax = sns.histplot(data=df['var_1'])
ax.set_xlabel('Variable')
ax.set_ylabel('frecuencia');
```

```python
ax = sns.histplot(data=df['var_1'], stat='density')
ax.set_xlabel('variable')
ax.set_ylabel('densidades de probabilidad ')
```

```python
ax = sns.histplot(data=df['var_1'], stat='density', kde=True, color='green')
ax.set_xlabel('variable')
ax.set_ylabel('densidades de probabilidad ')
```

```python
ax = sns.histplot(data=df['var_1'], stat='density', kde=True, bins=100, color='green')
ax.set_xlabel('variable')
ax.set_ylabel('densidades de probabilidad ')
```

```python
ax = sns.histplot(data=df['var_1'], stat='density', kde=True, bins=5, color='green')
ax.set_xlabel('variable')
ax.set_ylabel('densidades de probabilidad ')
```

```python
ax = sns.displot(df['var_1'], kind='kde')
```

```python
ax = sns.violinplot(data=df['var_1'])
ax.set_xlabel('variable')
ax.set_ylabel('valores')

```
