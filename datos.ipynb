```python
import pandas as pd
import random
import numpy as np
```

```python
# números  entre 15-18
normal_data = np.random.uniform(15, 18, 20)

# Generar 10 valores atípicos (outliers) 
outliers = np.random.uniform(19, 25, 10)  # Valores más altos como atípicos

# Combinar datos normales y atípicos
data = np.concatenate([normal_data, outliers])

# Mezclar aleatoriamente los datos
np.random.shuffle(data)

# Crear DataFrame
df = pd.DataFrame(data, columns=['Edad'])

# Mostrar estadísticas básicas para verificar la distribución
print("Estadísticas básicas:")
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear el gráfico de distribución
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Edad', bins=15)

# gráfico
plt.title('Distribución de Frecuencias de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()

# Calcular cuartiles y mediana
Q1 = df['Edad'].quantile(0.25)
Q2 = df['Edad'].median()  # La mediana es el segundo cuartil
Q3 = df['Edad'].quantile(0.75)

# Calcular rango intercuartil (IQR)
IQR = Q3 - Q1

# Mostrar resultados
print(f"Primer Cuartil (Q1): {Q1:.2f}")
print(f"Mediana (Q2): {Q2:.2f}")
print(f"Tercer Cuartil (Q3): {Q3:.2f}")
print(f"Rango Intercuartil (IQR): {IQR:.2f}")


plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Edad'])
plt.title('Edades')
plt.show()
```

1. Análisis Estadístico Descriptivo
Basado en los resultados mostrados, podemos inferir:

1.1 Medidas de Tendencia Central
La distribución muestra una clara bimodalidad, con dos grupos distintos:
Grupo principal: Estudiantes entre 15-18 años (n=20)
Grupo secundario: Valores atípicos entre 19-25 años (n=10)
1.2 Análisis de Dispersión
El IQR (Rango Intercuartil) indica la dispersión del 50% central de los datos
Los valores atípicos (>19 años) se identifican claramente en el boxplot como puntos individuales
2. Patrones Identificados
2.1 Distribución
El histograma muestra una distribución no normal (asimétrica)
Existe una clara separación entre la población principal y los valores atípicos
2.2 Clustering Natural
Se observan dos clusters naturales:
Cluster mayoritario: Estudiantes en edad típica
Cluster minoritario: Estudiantes con edad superior
