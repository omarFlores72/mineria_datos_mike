```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('publicidad.csv')
df
```

```python
x = df.iloc[:,:3].to_numpy()
y = df['ventas'].to_numpy().reshape(-1,1)

print(x.shape)
print(y.shape)
```

```python
mod_lin_multiple = LinearRegression()
mod_lin_multiple.fit(x,y)
```

```python
beta_0 = mod_lin_multiple.intercept_[0]
beta_tv = mod_lin_multiple.coef_[0][0]
beta_rd = mod_lin_multiple.coef_[0][1]
beta_im = mod_lin_multiple.coef_[0][2]

print('parametros: ')
print(f"\tbeta_0: {beta_0}")
print(f"\beta_tv: {beta_tv}")
print(f"\beta_im: {beta_im}")





```
