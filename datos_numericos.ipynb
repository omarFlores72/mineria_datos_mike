```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("etfs.csv")
df
```

```python
#grado de relacion entre QQQ Y USO
df['QQQ'].corr(df['USO'])
```

```python
#calculamos al grado de relacion entre XLY y XLP
df['XLY'].corr(df['XLP'])
```

```python
#Matriz de correlacion para todas las variables numericas
df.corr()
```

```python
#mapa de color (heradmap) para visualizar la matriz de correlacion
ax = sns.heatmap(df.corr(), vmin=-1, vmax=1, cmap="mako");
```

```python
#graficas de dispersion
ax = sns.scatterplot(data=df, x='XLF', y='GLD')
coef = df['XLF'].corr(df['GLD'])
plt.text(-.6, 6, f'Coeficiente de correlacion de pearson = {coef:.2f}')
```

```python
ax = sns.scatterplot(data=df, x='QQQ', y='XLK')
coef = df['QQQ'].corr(df['XLK'])
plt.text(-.3, .5, f'Coeficiente de correlacion de pearson = {coef:.2f}')
```

```python
ax = sns.scatterplot(data=df, x='VXX', y='SPY')
coef = df['VXX'].corr(df['SPY'])
plt.text(-20,-4, f'Coeficiente de correlacion de pearson = {coef:.2f}')
```

```python
#matriz de dispersion
pd.plotting.scatter_matrix(df, figsize=(12,12));
```

```python
#mapa hexagonal

ax = sns.jointplot(x=df['QQQ'], y=df['XLK'],
                kind="hex",marginal_kws=dict(bins=20))
```

```python
ax = sns.kdeplot(x=df['QQQ'], y=df['XLK'], fill=True)
```
